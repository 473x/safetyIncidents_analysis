# Introduction

This Python notebook analyses online patient feedback about acute hospitals in England. The analysis uses natural language processing to create a measure of 'patient-reported safety incidents' in the textual feedback. This measure is compared to staff-reported safety incidents and hospital-level mortality rates. 

Four research questions are addressed:

**RQ1: Is the automated patient-reported safety incident measure valid?** <br>
We test the reliability of the measure against manual coding of 2,000 high/low scoring items of patient feedback.


**RQ2: Does the high-scoring online feedback include potentially unnoticed and unresolved safety incidents?** <br>
We use manual coding of the patient feedback to identify whether the safety incidents being reported are part of formal hospital investigations and whether the issues raised remain unresolved.

**RQ3: Is the automated measure of patient-reported safety incidents independent of staff-reported safety incidents?** <br>
We examine correlations between our measure of patient-reported safety incidents and staff-reported safety incidents.

**RQ4: Does the measure of patient-reported safety incidents predict hospital-level mortality rates?** <br>
We use a linear mixed effects model to predict hospital-level mortality rates controlling for year, geographic region, hospital size, feedback wordcount and feedback sentiment.

The patient feedback data comes from [Care Opinion](https://www.careopinion.org.uk/) and the secondary data comes from [NHS Digital](https://digital.nhs.uk/).

## Measuring Sentiment

To measure the sentiment of the patient reviews we use the [VADER](https://github.com/cjhutto/vaderSentiment) (Valence Aware Dictionary and sEntiment Reasoner) algorithm for Python [(Hutto & Gilbert, 2014)](https://www.aaai.org/ocs/index.php/ICWSM/ICWSM14/paper/download/8109/8122). It is a widely used lexicon and rule-based tool designed to measure sentiment in social media. VADER takes account of common negations (e.g., 'not good'), intensity (e.g., 'very', '!!!!', ALL CAPS), slang (e.g., 'sux', 'lol') and emojis. VADER scores each word in context and then produces a composite score ranging from -1 (most negative) to +1 (most positive).

The following cell illustrates how VADER scores six test sentences.

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

test_sentences = ["Superb excellent ++",
                  'easy car park',
                  'he was not happy', 
                  'she was dying', 
                  'he died very happy',
                  'he died']

for sentence in test_sentences:
    score = analyzer.polarity_scores(sentence)
    print("Sentiment score: {:<6}  '{:}'".format(str(round(score['compound'],2)), sentence))
```

```python
# create sentiment extension
def get_sentiment(doc):
    # can return pos, neg, neutral, and overall/compound
    return sentiment_analyzer.polarity_scores(doc.text)['compound']

# create custom extension
Doc.set_extension('sentiment', getter=get_sentiment, force=True)

# function to get sentiment
def getDocSentiment(doc):
    '''
    Uses VADER to get normalized weighted composite score for each sentence
    Output: composite sentiment float (-1 to 1)
    '''
    sentence_sentiments = []
    for sent in doc.sents:
        sentiment = round(sentiment_analyzer.polarity_scores(sent.text)['compound'], 3)
        sentence_sentiments.append(sentiment)
    try:
        doc_sentiment_sentence_mean = round(sum(sentence_sentiments)/len(sentence_sentiments), 3)
    except:
        doc_sentiment_sentence_mean = None 
          
    return doc_sentiment_sentence_mean
    
```

```python
# apply the VADER sentiment algorithm to the year-trust aggregated patient feedback
df['stories_sentiment'] = df.progress_apply(lambda x: getDocSentiment(x['stories_doc']), axis=1)
```

## Measuring Patient-Reported Safety Incidents

The traditional approach to textual analysis has been to count the frequency of words and phrases associated with the target concept (e.g., ‘died’, ‘dying’, ‘near miss’, ‘avoidable’). This approach has the benefit of being transparent (one knows exactly which words are counted and one can examine these words in context). However, this approach is blunt because it weights each word equally (i.e., ‘1’ for present and ‘0’ for absent), which can lead to an under-weighting of low-frequency high-validity words relative to high-frequency low-validity words (e.g., ‘died’ may be infrequent relative to ‘avoidable’).

Supervised learning is a more recent approach that models statistical patterns in manually coded data to create a classification algorithm. This approach has the benefit of replicating human coding. However, these models often lack transparency (i.e., it can be difficult to discern why certain text was or was not classified) and the quality of these models is limited by the quality of the input data (e.g. if the human judgements are unreliable, as is the case with staff-reported safety incidents, then the supervised classifier will be unreliable)

Accordingly, we chose to measure patient-reported safety incidents following the Distributed Dictionary Representation (DDR) methodology described by [Garten et al. (2018)](https://doi.org/10.3758/s13428-017-0875-9). This method has the benefit of not being skewed by low-relevance but high-frequnecy terms (like word frequency methods), it is tolerant of the spelling mistakes that are common online (unlike word frequency methods), and it produces normally distributed scores (unlike word frequency methods and supervised learning methods that categorise text).

The DDR methodology requires a distributed language model, where each word within the language model is represented by a vector, or word embedding, so that the similarity of a text to a target concept can be measured. We use the [large spaCy english language model 2.3.1](https://github.com/explosion/spacy-models/releases//tag/en_core_web_lg-2.3.1). This language model has 684,831 unique vectors (i.e. tokens or words) represented in 300 dimensions. Each word in the language model is represented by a vector of 300 numbers. The relationship between any two words, or vectors, in the language model is encoded numerically in the relation between their respective vectors.

The following cell illustrates how the DDR methodology works. The target concept is the single target term 'died'. Using the same sentences as used to illustrate the measurement of sentiment, we then score the similarity of each sentence to the target target term. A score of '1' indicates maximum similarity.

```python
target_term = 'died'

# compare target_term to each test sentence
for sentence in test_sentences:
    score = np.round(nlp(sentence).similarity(nlp(target_term)), 2)
    print("Similarity to 'died':  {:<6}  '{:}'".format(str(score), sentence))
```

### Step 2: Increase candidate target terms

```python
# function to generate target terms
def getTokensNearVector(nlp, string_terms, limit=300):
    '''
    Takes a string of words and uses the language model to suggest similar words
    '''
    doc = nlp(string_terms)
    
    vectors = []
    for tok in [tok for tok in doc if tok.is_alpha]:
        try:
            vectors.append(tok.vector)
        except:
            print('error getting vector for ', tok.text)
    
    #sum vectors
    vector_sum=0
    for v in vectors:
        vector_sum=vector_sum+v
    #get mean vector
    vector=vector_sum/len(vectors)
    
    # Format the input vector for use in the distance function
    p = [np.array(vector)]
    # Format the vocabulary for use in the distance function
    ids = [x for x in nlp.vocab.vectors.keys()]
    vectors1 = [nlp.vocab.vectors[x] for x in ids]
    vectors2 = np.array(vectors1)
    
    # measure distances
    dist = distance.cdist(p,vectors2)[0]
    
    # create df of results
    df = pd.DataFrame({'distance':[x for x in dist]})
    df.sort_values('distance', inplace=True)

    # collect results
    words = []
    for i in df.index[:limit]:
        word_id = ids[i]
        output_word = nlp.vocab[word_id].text
        output_word = output_word.lower()
        if output_word not in words:
            words.append(output_word.lower())
            
    # remove input tokens
    words = [word for word in words if word not in [tok.lower_ for tok in doc]]
    
    # output
    return words
```

```python
# use the distributed language model to get words similar to the initial target terms
related_targetTerms_v1 = []
for term in initial_targetTerms.split():
    term_list = getTokensNearVector(nlp, term, limit=100)
    related_targetTerms_v1 = related_targetTerms_v1+[x for x in term_list if x not in related_targetTerms_v1]
related_targetTerms_v1 = ' '.join(sorted(set(related_targetTerms_v1+initial_targetTerms.split())))   
related_targetTerms_v1
```

### Step 3: Assessing the validity high-scoring text

High scoring sentences for each candidate target term are examined. The aim is to select target terms that consistently capture the most severe safety incidents.

```python
# function to get high-scoring sentences
def getSimilarSentencesForEachToken(nlp, docs, target_terms, n_sentences=10, 
                                    min_sentence_length=6):
    '''
    Finds most similar sentences for each tonken and all tokens together
    '''
     
    target_terms = nlp(target_terms.strip())
    
    # get highest scoring sentences
    for tok in target_terms:
        sentences = []
        similarities = []
        if tok.has_vector:
            #print('TOKEN = ', tok.text)
            for doc in docs:
                for sent in doc.sents:
                    if sent.has_vector:
                        if len(sent)>min_sentence_length:
                            sentences.append(sent.text)
                            similarities.append(tok.similarity(sent))
    
        df_temp=pd.DataFrame({tok.text: similarities,
                              'sentence': sentences}).round(3).copy()
        #print
        with pd.option_context('display.max_colwidth', 0):
            display(df_temp.sort_values(tok.text, ascending=False).head(n_sentences)
                    .style.set_properties(**{'text-align': 'left'}).hide_index())
```

### Step 4: Assessing the reliability of target-terms

Here we apply each candidate target term to the reviews individually, and then examine the correlations between the measures. The aim is to identify target terms that have good inter-term reliability.

```python
def reliabilityOfDDR(nlp, docs, target_terms):
    '''
    Finds the Cronbach alpha for target-terms
    '''
    
    target_words = nlp(target_terms)
    
    # score data
    docs_measures = []
    for doc in docs:
        if doc.has_vector:
            doc_measures = {}
            out_of_vocab = []
            doc_measures['doc'] = doc
            for target_token in target_words:
                if target_token.is_oov:
                    pass
                else:
                    doc_measure = target_token.similarity(doc)
                    doc_measures[target_token] = doc_measure
            docs_measures.append(doc_measures)
        else:
            pass
        
    # round    
    df_temp = pd.DataFrame(docs_measures).round(3)     
    # drop doc
    try:
        df_temp.drop('doc', axis=1, inplace=True)
    except:
        pass
    
    # correlations
    df_corr = df_temp.corr()
    display(df_corr.style.background_gradient(cmap='Blues', axis=None).set_precision(2))
        
    # reliability analysis
    shape1=df_temp.shape
    df=df_temp.dropna(axis='index').copy()
    shape2=df.shape
    if shape1 != shape2:
        print('\n\nDropped rows with NAs: ', shape1[0]-shape2[0])
    
    ca = pingouin.cronbach_alpha(data=df, ci=0.95)
    print('Cronbach Alpha = ', round(ca[0], 3), '    95% CI = ', ca[1])
```

### Apply measure to patient feedback

```python
def createDDR(nlp, df, doc_variable, target_terms, variable_to_create):
    '''
    creates a DDR variable from a spacy doc based on target_terms
    '''
    
    target_terms_doc = nlp(target_terms)
    
    similarity_scores = []
    for doc in df[doc_variable].values:
        if doc.has_vector: 
            similarity_scores.append(target_terms_doc.similarity(doc))
        else:
            similarity_scores.append(None)
    df[variable_to_create] = similarity_scores
```

```python
createDDR(nlp, df, 'stories_doc', safetyIncident_targetTerms, 'ddr_safetyIncident')
```

```python
# function to print multi-year correlations in plot
def corrfunc(x, y, **kws):
    if (len(x)>1) & (len(y)>1):
        r, p = stats.spearmanr(x, y)
    else: # if there is no data
        r = 0
        p = 1
    ax = plt.gca()
    # count how many annotations are already present
    n = len([c for c in ax.get_children() if 
                  isinstance(c, matplotlib.text.Annotation)])
    pos = (.25, .8 - .1*n)
    
    # convert r to conditional string to create extra space to counterbalance '-' neg values
    r = str(round(r, 2)) if r <0 else ' '+str(round(r,2))
    r = r.replace('0.', '.')
            
    if p < 0.001:
        ax.annotate("{}:  r = {}***".format(kws['label'], r), xy=pos, xycoords=ax.transAxes).set_weight('bold')
    elif p < 0.01:
        ax.annotate("{}:  r = {}**".format(kws['label'], r), xy=pos, xycoords=ax.transAxes).set_weight('bold')
    elif p <= 0.05:
        ax.annotate("{}:  r = {}*".format(kws['label'], r), xy=pos, xycoords=ax.transAxes).set_weight('bold')
    elif p==1: # if there is no data
        ax.annotate("{}: No data".format(kws['label']), xy=pos, xycoords=ax.transAxes)
    else:
        ax.annotate("{}:  r = {}".format(kws['label'], r), xy=pos, xycoords=ax.transAxes)
```

## RQ4: Predicting hospital-level mortality


In this section, we test whether staff-reported safety incidents and the measure of patient-reported incidents were predictive of hospital-level mortality taking account of multiple controls. The unit of analysis was a hospital-year (N = 792), representing 134 distinct hospitals (from a population of 139) over the period from 2013 to 2019 (this missing hospital-years was due to trust-years with less than 10k words of patient feedback being excluded).

```python
# due to the skewed distribution of staff-reported safety incidents (all), we convert to a log score
df['si_all_rate_log'] = np.log(df['si_all_rate']+0.01)
df.hist(bins=60, layout=(1,2), column=['si_all_rate', 'si_all_rate_log'], figsize=(10, 2));
```

```python
# due to the skewed distribution of staff-reported safety incidents (severe), we convert to a log score
df['si_severe_rate_log'] = np.log(df['si_severe_rate']+0.01)
df.hist(bins=60, layout=(1,2), column=['si_severe_rate', 'si_severe_rate_log'], figsize=(10, 2));
```

A linear mixed effects model was constructed using year (2013-2019 inclusive) and region (North, Midlands, London, South) as crossed random effects based on exploratory analysis indicating differences in some variables across these factors. The model also controlled for the total number of hospital spells, total word count in the patient feedback, and a text-based measure of overall positive/negative sentiment in the patient feedback (fixed effects). All numeric explanatory variables were linearly normalized to a range of 0 to 1, i.e. so that the minimum value in the data was transformed to 0 and the maximum value to 1.

```python
# create mixed linear model
target = ['trust_shmi']
controls = ['trust_spells', 'wordcount', 'stories_sentiment']
predictors = ['si_all_rate_log', 'si_severe_rate_log', 'ddr_safetyIncident']

# order
vars_in_order = controls+predictors
vars_in_order.reverse()

# normalize data 0-1 for coeffient plotting
df_norm =df[['trust_id', 'year', 'trust_region']+vars_in_order+target].copy()
df_norm.dropna(inplace=True)
vars_to_normalize = vars_in_order

for var in vars_to_normalize:
    df_norm[var] = (df_norm[var]-df_norm[var].min())/(df_norm[var].max()-df_norm[var].min())

# create formula
formula = target[0]+' ~ '+' + '.join(vars_in_order)
print(formula)

# create grouping varibales
vcf = {"trust_region": "0 + C(trust_region)", "year": "0 + C(year)"}

# add an artificial group
df_norm['group'] = 1

# model
model = sm.MixedLM.from_formula(formula, df_norm, re_formula= '0', vc_formula = vcf, groups = 'group')
result = model.fit()
result.summary()
```

Neither all staff-reported safety incidents (si_all_rate_log) nor severe staff-reported safety incidents (si_severe_rate_log) were significantly predictive of hospital-level mortality. An increase of 0.1 in the all-incident measure was associated with a decrease of 0.001 in hospital-level mortality (95% CI = -0.047, 0.046; z = -0.02; p = .98), holding all other variables fixed. Similarly, an increase of 0.1 in the severe-incident measure was associated with a decrease of 0.011 in hospital-level mortality (95% CI = -0.050, 0.027; z = -0.58; p = 0.56), holding all other variables fixed.

The measure of patient-reported safety incidents was found to be significantly predictive of hospital-level mortality. An increase of .1 in this measure was associated with an increase of 0.085 in hospital-level mortality (95% CI = 0.042, 0.128; z = 3.87; p < .001). An increase of 0.01 in hospital-level mortality translates to an increase of 1% in actual deaths relative to an expected number of deaths based on characteristics of the patients treated at a given hospital in a given year.

```python
def coefPlot(result, rename={}, figsize=(6, 3), save_file = None):
    coefs = result.params
    coefs.rename(rename, inplace=True)
    coefs.drop(labels = ['Intercept', 'year Var', 'trust_region Var'], inplace = True)
    ci = result.conf_int()
    errors = (ci[1] - ci[0]) / 2
    errors.rename(rename, inplace=True)
    errors.drop(labels = ['Intercept', 'year Var', 'trust_region Var'], inplace = True)
    plt.clf()
    plt.figure(figsize=figsize)
    plt.errorbar(x = coefs, y = coefs.index, xerr = errors, fmt = 'o', zorder=2, ecolor='black', color='black')
    plt.axvline(x = 0, color = 'black', zorder=1)
    plt.title(f'Scaled parameter estimates with 95% confidence intervals\n')
    if save_file != None:
        plt.savefig(save_file, bbox_inches='tight');
```

```python
# rename variables
rename_cols = {'trust_spells': 'Hospital spells',
               'wordcount': 'Wordcount of patient reviews',
               'stories_sentiment': 'Patient review sentiment',               
               'si_all_rate_log': 'Staff-reported incidents (all)',
               'si_severe_rate_log': 'Staff-reported incidents (severe)',
               'ddr_safetyIncident': 'Patient-reported safety incidents'}

coefPlot(result, rename_cols, figsize=(8, 3));
plt.savefig('fig_parameter_estimates.svg', format='svg', dpi=1200, bbox_inches='tight')
```

**Interpretation**

**Hospital spells.** As the patient-throughput increases, the Summary Hospital-level Mortality Indicator (SHMI) decreases (i.e., there are fewer deaths than expected). This suggests that large hospitals have lower than expected mortality rates.

**Patient feedback wordcount.** As patients write more and longer reviews, SHMI increases. This suggests that patients have more to say about hospitals with higher than expected mortality rates (controlling for the patient throughput).

**Patient feedback sentiment.** There is a tendency for increased sentiment (i.e., more positive text) to be a associated with decreased SHMI.

**Staff-reported safety incidents (all & severe).** There is no discernible association between staff-reported safety incidents (either all or severe) and hospital-level mortality rates.

**RQ4: Patient-reported safety incidents.** There is an association between patients reporting safety incidents and an increase in SHMI that is independent of feedback sentiment. This suggests that beyond merely expressing sentiments, patients are reporting information that might inform high hospital-level mortality rates.
    

# Robustness checks


According to [McKenny et al (2018)](https://doi.org/10.1177%2F0149206316657594), natural language processing techniques are associated with various types of error, including:
1. 'Transient error' due to variations in language use over time.
2. 'Specific factor error' due to choices made when compiling word lists.
3. 'Algorithm error' due to variations in how algorithms operate.




## Transient error
Temporal variation in language use is inherent in human language and cannot be eliminated. Instead, it needs to be investigated and controlled for. The current analysis controls for transient error, by having year as a control variable. However, this control, does not provide much insight into transient error. In this section, we explore transient error in the patient-reported safety incident measure.

### Test-retest reliability

Following [McKenny et al (2018)](https://doi.org/10.1177%2F0149206316657594), we quantify transient error by calculating the correlation coefficients for the safety incident measure for each consecutive pair of years, and averaging the correlation coefficients. 

```python
# get sorted list of years
years = sorted(list(set(df['year'])))
# create consecutive pairs
year_pairs = list(zip(years[:-1], years[1:]))
list(year_pairs)
```

```python
# create wide df
df_wide = df[['year', 'trust_id', 'ddr_safetyIncident']].pivot(index='trust_id', columns='year')
df_wide.columns = df_wide.columns.droplevel()
df_wide.head(3)
```

```python
correlations = []
for pair in year_pairs:
    correlation = df_wide[list(pair)[0]].corr(df_wide[list(pair)[1]]).round(3)
    correlations.append(correlation)
    print(pair, '  Pearson corr =  ', correlation)
```

```python
# mean correlation
sum(correlations)/len(correlations)
```

Correlating the safety incident scores for each pair of consecutive years revealed potentially substantial and consistent transient error. The test-retest correlations ranged from 0.408 to 0.579, with a mean correlation of 0.459. Thus up to 54% (1-mean correlation) of the variance is due to transient error.

[McKenny et al (2018)](https://doi.org/10.1177%2F0149206316657594) report comparable rates of transient error and note that measures of transient error can be artificially high if there are changes in the underlying construct (e.g., hospital improvements). The violin plots of changes in sentiment and patient-reported safety incidents over time suggest that there may be changes in sentiment and safety incidents over time. This could be due to real changes in the phenomena, or changes in language use.

The implications of this transient error are: (1) it is necessary to control for changes over time, (2) the language model and target terms should be kept up-to-date to capture changes in language use.


## Specific factor error

There are necessarily many decisions required to do natural language processing (e.g., how text is cleaned and prepared, the choice of key words and concepts to explore, and how specific concepts are measured). Our analysis entailed minimal pre-processing of the texts (we did not remove or change any of the original text). However, we made many decisions when choosing our target terms. Although we followed a clearly laid out 4-step methodology (described above), different choices of target terms would have changed the results, and perhaps the overall finding. In this section we assess the extent to which the findings may be an byproduct of our choice of target terms.

To assess the extent to which the findings were an byproduct of the chosen target terms, we re-ran the mixed model with (1) a random selection of half of our target terms and (2) the rejected target terms.

## Algorithm error

Algorithms entail many choices. Even an algorithm for counting the frequency of a word in a text entails choices about how to handle capitalization, prefixes, suffixes, root forms (lemmas) and misspellings. An algorithm that uses word embeddings (as we did) has additional choices about the data selected to train the algorithms on. To asses the extent to which the findings may be the result of algorithm error, the analysis was re-run using a custom language model.


### Using a custom language model

Our main study used a large, robust, and publicly available language model [(spaCy, 2020)](https://github.com/explosion/spacy-models/releases//tag/en_core_web_lg-2.3.1) so as to test our predictions with as few degrees of freedom as possible. However, it is widely recognised that such generic models are sub-optimal in specific cases. Biases can be introduced into these generic models through peculiar statistical patterns in the input data. The language model we used was trained on [OntoNotes](https://catalog.ldc.upenn.edu/LDC2013T19), which contains a many news articles and broadcasts from 2013, and [GloVe Common Crawl](https://nlp.stanford.edu/projects/glove/) which contains a lot of webpage text. These training datasets are, arguably, quite far removed from patient feedback that is very domain specific.

To check whether there was a bias in the language model that materially impacted our findings, we trained our own language model. Because the patient feedback of acute trusts only comprised about 20 million words, and it is recommended to have at least 100 million words to build a language model, we expanded our corpus of patient feedback to include all the feedback available via the Care Opinion API (i.e., including feedback on non-acute NHS organisations). This yielded a dataset of over 400k items of patient feedback, comprising 58 million words. Although still small, this dataset is highly tailored to our interest and thus deemed viable to create a custom language model. The language model was created using the Word2Vec algorithm, yielding a model with 54,337 unique tokens in 300 dimensions.
